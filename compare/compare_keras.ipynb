```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# compare : 0 => A > B / 1 => A < B / 2 => A=B
data = pd.read_csv("./compare_data.csv", index_col=0, dtype=int)
data
```

```python
x_train, x_test, y_train, y_test = train_test_split(data[["A", "B"]], data["compare"], stratify=data["compare"],
                                                    random_state=0)
del data
```

```python
from keras.models import Sequential

from keras.layers import Dense, InputLayer, Activation, Normalization
from keras.activations import relu, softmax

from keras.losses import sparse_categorical_crossentropy
from keras.optimizers import Adam
from keras.metrics import sparse_categorical_accuracy

import datetime
from tensorflow.python.keras.callbacks import TensorBoard

%load_ext tensorboard
```

```python
model = Sequential([
    InputLayer(2),
    Normalization(),
    Dense(10),
    Activation(relu),
    Dense(10),
    Activation(relu),
    Dense(3),  # 클래스의 수에 맞게 조정
    Activation(softmax)
])

model.compile(
    loss=sparse_categorical_crossentropy,
    optimizer=Adam(learning_rate=0.01),
    metrics=[sparse_categorical_accuracy]
)

tensorboard = TensorBoard(log_dir="logs/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))

model.summary()
```

```python
model.fit(x_train, y_train, validation_split=0.3, epochs=10, verbose=0, callbacks=[tensorboard])
```

```python
%tensorboard --logdir logs
```

```python
model.evaluate(x_test, y_test)
```

```python
import time

# 시간 측정 시작
start_time = time.time()

# 특정 반복문
for data in zip(x_train["A"], x_train["B"]):
    if data[0] > data[1]:
        print(">")
    elif data[0] < data[1]:
        print("<")
    else:
        print("=")

# 시간 측정 종료
end_time = time.time()

# 경과 시간 출력
elapsed_time = end_time - start_time
print(f"경과 시간: {elapsed_time} 초")
```
